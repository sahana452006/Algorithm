```python
import time
import numpy as np
import matplotlib.pyplot as plt
def binary_search(arr, target, left, right):
    if left <= right:
        mid = (left + right) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] > target:
            return binary_search(arr, target, left, mid - 1)
        else:
            return binary_search(arr, target, mid + 1, right)
    return -1
def measure_time(n):
    arr = np.arange(n)
    target = np.random.randint(0, n)
    start_time = time.time()
    binary_search(arr, target, 0, len(arr) - 1)
    end_time = time.time()
    return end_time - start_time
n_values = np.arange(100, 2000, 100)
time_taken = []
for n in n_values:
    time_taken.append(measure_time(n))
plt.figure(figsize=(8, 6))
plt.plot(n_values, time_taken, label='Time taken for Binary Search', color='blue')
plt.xlabel('Number of Elements (n)', fontsize=12)
plt.ylabel('Time Taken (seconds)', fontsize=12)
plt.title('Binary Search TimeComplexity (Recursive)', fontsize=14)
plt.grid(True)
plt.legend()
plt.show()
```
